# 第一课 数字冰壶比赛简介

## 1.1 冰壶比赛简介

冰壶运动是在以队为单位冰上进行的投掷性竞赛，被誉为“冰上的国际象棋”，它考验参与者的体能与脑力，展现动静之美，取舍之智慧，属于冬奥会比赛项目。

<center><img src="images/CurlingThrow.gif"></center>

简单来说，冰壶比赛就是2个队，每队4人，轮番把16个40斤的冰壶投进约摸30米远的大本营中（一人投俩）。比赛结束后，在大本营中，所有冰壶更靠近圆心的队伍得分，并且有多少冰壶更靠近圆心就得多少分。

<center><img src="images/CurlingScore.png" width=400></center>

就这样进行10场比赛，每场比赛得分加起来作为最后的总得分。在冰壶比赛中，由于后投者可以将大本营中已有的冰壶撞开，所以这并不是一个简单的比准头的比赛，而是一场讲究策略的博弈。

## 1.2 数字冰壶比赛平台

数字冰壶比赛平台实现了冰壶运动在数字空间中的呈现，学生可以在虚拟空间中采用数字模拟器自行训练AI选手，设计各种策略实现数字空间中冰壶的智能投掷，还可以和其它学生设计的AI选手进行数字比赛对抗，深入理解强化学习、博弈论等理论并在实验中验证。

数字冰壶比赛平台服务器端程序基于Unity开发，用作数字冰壶比赛的游戏引擎。该平台提供AI对战以及人机对战功能，平台与选手编写的AI程序采用TCP协议进行通信，参赛选手编写的AI程序需要根据接收的消息制定出决策，并调用发送接口发送给对战平台。

### 1.2.1 数字冰壶比赛服务器

点击页面左上角Jupyter菜单中的[Run]菜单项，点击该该菜单项的[Start Curling Server]子菜单项，即可启动一个数字冰壶比赛服务器。

<center><img src="images/StartServer.png" width=600px></center>

数字冰壶比赛服务器会在新开浏览器页面中启动，第一次加载速度较慢，请耐心等待。

数字冰壶比赛服务器主界面如下图所示，该服务器界面支持汉语、英语、俄语、法语、西班牙语、阿拉伯语、日语、韩语、孟加拉语、乌尔都语共十国语言，主界面中标注①的部分就是语言选择接口。

主界面中标注②的部分是比赛模式选择接口，共提供[四局制]（初赛模式）、[八局制]（决赛模式）、[无限局制]（训练模式）、[人机对战]（游戏模式）四种模式。

主界面右下角标注③的部分给出了数字冰壶比赛服务器的连接信息，具体包括 ConnectKey（服务器连接密钥）、IP（服务地址）和 Port（服务端口）。<b>每次启动服务器该信息中的具体数据都会有变化</b>，这些数据在后续的程序代码编写中会用到。

#### 数字冰壶比赛服务器的人机对战模式

下面以人机对战为例，展示数字冰壶比赛服务器的功能。

在数字冰壶服务器的主界面中点击【人机对战】按钮，即可进入如下图所示的对战准备界面。

<center><img src="images/CurlingServerV.png"></center>

在人机对战界面的左上角，可以输入人类玩家姓名、选择人类玩家先手/后手、选择对战的AI。可以选择的对战AI包括[笨笨机器人]（只会把冰壶投出界的陪练AI、以往两届数字冰壶比赛的四强AI、[我的机器人]（数字冰壶比赛课程平台提供的基础AI）。

输入姓名、保持默认的人类玩家先手、选择默认的[笨笨机器人]作为对战AI，依次点击【启动AI】>>【准备】>>【开始对战】几个按钮，即可进入如下图所示的对战界面。

<center><img src="images/CurlingHMvsAI1.png"></center>

该界面中的右上方有一块“人类玩家控制面板”，面板中有三个滑动条可以用来调节数字冰壶投掷的三个参数：冰壶投掷时的初速度v0 ( **0 ≤ v0 ≤ 6** ）、冰壶投掷时的横向偏移h0 ( **-2.23 ≤ h0 ≤ 2.23** )、冰壶投掷时的初始角速度ω0 ( **-15.7 ≤ ω0 ≤ 15.7** )。

界面中下方有一个用于示意投掷参数用途的冰壶，如果投掷参数有调整，该冰壶的横向位置会随h0变动，并且指示初速度和旋转角速度的箭头也会随v0和ω0变动。这三个投掷参数的默认值为v0=3、h0=0、ω0=0，暂且保持默认值不变，点击【发射】，冰壶即会被投出，在冰壶场地中滑行，并最终停止在场地上。如下图所示。

<center><img src="images/CurlingHMvsAI2.png"></center>

从上图可以看出，v0=3、h0=0、ω0=0这组参数正好可以把冰壶投掷到大本营的靶心附近。由于数字冰壶比赛服务器的运动模型中引入了小范围的随机摩擦，所以同样的参数投掷出去的冰壶，位置会有少许不同。这个设定为比赛引入了一些偶然性，也更接近客观世界的实际冰壶比赛。

人类玩家投掷的冰壶停止移动后，可以看到[笨笨机器人]投掷的冰壶在场地最右侧快速滑出了界外，这就是[笨笨机器人]的职责——只负责陪练。界面右上角的四个倍速按钮可以改变冰壶投掷过程的渲染速度，由于正式比赛时都是采用[四倍速]，因此日常测试也建议采用[四倍速]。

数字冰壶比赛和实际的冰壶比赛一样，每局中双方各有8个冰壶，在数字冰壶比赛服务器对战界面的左上方显示了双方剩余的冰壶数量。在“人类玩家控制面板”中调整投掷参数，将剩余的7个冰壶都投掷出去。双方的8个个冰壶都投掷结束后，会进入如下图所示的界面。

<center><img src="images/CurlingHMvsAI3.png"></center>

界面中显示的是当前双方的比分，目前只进行了第一局，由于[笨笨机器人]只会投掷出界壶，所以比分上是人类玩家得分。

数字冰壶比赛的初赛阶段设定为每场比赛四局，决赛阶段设定为每场比赛八局，人机对战模式是按照决赛阶段八局制设定的。可以点击【下一局】按钮，即会开始新的一局对战。也可以点击【返回主菜单】按钮，选择其他模式，开始真正的学习进程。

### 1.2.3 数字冰壶比赛平台通讯协议

前面演示了数字冰壶比赛平台的对战模式，该模式中人类玩家可以在控制面板中通过简单的操作完成投掷。在实际比赛中参赛队伍需要编写python程序，采用TCP协议和数字冰壶比赛服务进行通信，按照通讯协议规约完成单个冰壶的投掷，并在数字空间中与另一个AI进行对抗，完成智能对战。

数字冰壶比赛平台常用通讯协议规约如下表所示。AI程序可以遵循此通讯协议和数字冰壶服务器进行通讯，获取得分区内的冰壶球位置，或者给出投掷指令，从而完成虚拟对战。

<center><img src="images/protocol.png"></center>

指令消息代码对应描述如下（详细使用参见课程《02.数字冰壶比赛中的Socket通信》）：

- CONNECTKEY =>：AI选手发送连接连接密钥（在数字冰壶比赛客户端右下角查看）；
- CONNECNAME <=：客户端通知连接信息。参数1-客户端给AI选手生成的临时选手名，Player1表示首局先手，Player2表示首局后手；
- ISREADY <=：客户端通知准备完毕；
- READYOK =>：AI选手发送准备完毕；
- NAME =>：AI选手发送选手名。参数1-AI选手名，如不发送此消息则沿用临时选手名；
- NEWGAME <=：客户端通知开始比赛；
- POSITION <=：客户端通知16个冰壶球的当前坐标，参数顺序同当前对局投掷顺序，（0，0）坐标表示未投掷或已出界的球；
- SETSTATE <=：客户端通知赛局状态。参数1-当前是第几壶（从0开始计数），参数2-当前是第几局（从0开始计数），参数3-比赛总对局数，参数4-预备投壶队伍（0为先手/蓝方，1为后手/红方）；
- GO <=：客户端通知AI选手可以进行投壶；
- BESTSHOT =>：AI选手给出投掷信息。参数1-冰壶初速度，参数2-冰壶投掷时的横向偏移，参数3-冰壶初始角速度；
- MOTIONINFO <=：客户端通知冰壶在冰壶球运动至赛道中间时的运动状态信息。参数1-当前运动冰壶x坐标，参数2-当前运动冰壶y坐标，参数3-当前运动冰壶速度x方向分量，参数4-当前运动冰壶速度y方向分量，参数5-当前运动冰壶旋转角速度；
- SWEEP =>：如AI选手选择擦冰则发送此消息，擦冰最长持续到对方前卫线。参数1-擦冰距离；
- SCORE <=：客户端通知本局比赛AI选手得分。参数1-本局比赛AI选手得分；
- TOTALSCORE <=：客户端通知总得分。参数1-持篮球选手总得分，参数2-持红球选手总得分；
- GAMEOVER <=：客户端通知比赛结果。参数1-胜利(WIN)、失败(LOSE)或平局(DRAW)。

> 备注：CONNECTKEY消息代码仅适用于课程平台，**比赛平台中不支持此消息代码**。

除以上常用通讯协议规约中涉及到的消息代码之外，数字冰壶比赛平台还支持特殊功能消息代码如下（详细使用参见课程《04.数字冰壶比赛中的对战策略》）：

- CENTERLINE_VIOLATION <=：客户端通知对手AI触犯中线规则，详见《04.数字冰壶比赛中的对战策略》；
- CENTERLINE_CHOICE =>：AI选手发送针对对手触犯中线违规的选择。参数1-RESET(场上冰壶复位到对手犯规前位置)或KEEP(场上冰壶保持当前位置)；
- RESETPOSITION：重设冰壶球坐标，参数和POSITION相同；
- RESETSTATE：重设赛局状态，参数和SETSTATE相同。

> 备注：RESETPOSITION和RESETSTATE消息代码仅适用于课程平台，且仅在客户端向AI选手发送了GO消息、等待AI选手的BESTSHOT消息的状态下有效，**比赛平台中不支持这两个消息代码**。
